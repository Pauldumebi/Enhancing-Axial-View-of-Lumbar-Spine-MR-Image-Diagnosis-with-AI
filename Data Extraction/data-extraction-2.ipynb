```python
import pandas as pd
import numpy as np
import os,shutil

df = pd.read_excel('Radiologist Report Analysis.xlsx')

df.head(3)
```

### Splitting images into respective folders

```python
# Replace missing values with 30
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(30, inplace=True)

cols = ['Patient ID', 'D3_CS', 'D4_CS', 'D5_CS']
#df[cols] = df[cols].applymap(np.int64)

# get the path/directory
folder_dir = "dataset/Composite_Images/"

def checkFileExistAndMove(id, Disc, status):
    if len(str(id)) == 1:
        id = '000' + str(id)
    elif len(str(id)) == 2:
        id = '00' + str(id)
    elif len(str(id)) == 3:
        id = '0' + str(id)
        
    print(id, 'id')

    image = 'C1_'+ id + '_' + Disc + '.png'
    path = folder_dir + image
    isExisting = os.path.exists(path)
    if isExisting:
        if status == 'No':
            original = path
            target = "dataset/No Issue/" + image
            shutil.move(original, target)
        elif status == 'Mild':
            original = folder_dir + image
            target = "dataset/Mild Issue/" + image
            shutil.move(original, target)
        elif status == 'Moderate':
            original = folder_dir + image
            target = "dataset/Moderate Issue/" + image
            shutil.move(original, target)
        elif status == 'Serious':
            original = folder_dir + image
            target = "dataset/Serious Issue/" + image
            shutil.move(original, target)

for i in range(1, 576):
    IDs = df['Patient ID'].tolist()
    if(i in IDs ):
        getRow = df.loc[df['Patient ID'] == i]
        if getRow['D3_CS'].to_string(index=False) == 'No':
            checkFileExistAndMove(i, 'D3', 'No')
        elif getRow['D3_CS'].to_string(index=False) == 'Mild':
            checkFileExistAndMove(i, 'D3', 'Mild')
        elif getRow['D3_CS'].to_string(index=False) == 'Moderate':
            checkFileExistAndMove(i, 'D3', 'Moderate')
        elif getRow['D3_CS'].to_string(index=False) == 'Serious':
            checkFileExistAndMove(i, 'D3', 'Serious')

        if getRow['D4_CS'].to_string(index=False) == 'No':
            checkFileExistAndMove(i, 'D4', 'No')
        elif getRow['D4_CS'].to_string(index=False) == 'Mild':
            checkFileExistAndMove(i, 'D4', 'Mild')
        elif getRow['D4_CS'].to_string(index=False) == 'Moderate':
            checkFileExistAndMove(i, 'D4', 'Moderate')
        elif getRow['D4_CS'].to_string(index=False) == 'Serious':
            checkFileExistAndMove(i, 'D4', 'Serious')

        if getRow['D5_CS'].to_string(index=False) == 'No':
            checkFileExistAndMove(i, 'D5', 'No')
        elif getRow['D5_CS'].to_string(index=False) == 'Mild':
            checkFileExistAndMove(i, 'D5', 'Mild')
        elif getRow['D5_CS'].to_string(index=False) == 'Moderate':
            checkFileExistAndMove(i, 'D5', 'Moderate')
        elif getRow['D5_CS'].to_string(index=False) == 'Serious':
            checkFileExistAndMove(i, 'D5', 'Serious')
```
